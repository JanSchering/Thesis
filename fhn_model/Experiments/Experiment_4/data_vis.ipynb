```python
import plotly.express as px
import pandas as pd
import numpy as np
import torch as t
```

```python
num_param_steps = 15
k1_vals = t.linspace(0.5,1,num_param_steps)
k2_vals = t.linspace(0,0.5,num_param_steps)
k3_vals = t.linspace(0,0.5,num_param_steps)
mesh_k1, mesh_k2, mesh_k3 = t.meshgrid(k1_vals, k2_vals, k3_vals)
distances = np.load("distances_ex9_3.npy")
```

```python
data_stack = np.vstack((mesh_k1.flatten().numpy(), mesh_k2.flatten().numpy(), mesh_k3.flatten().numpy(), distances.flatten(), np.log(distances).flatten()))
data_stack.T.shape
```

```python
df = pd.DataFrame(data_stack.T, columns=["k1", "k2", "k3", "distance", "log_distance"])
```

```python
df.head()
```

```python
px.defaults.color_continuous_scale = px.colors.sequential.Viridis
fig = px.scatter_3d(df,
                    x='k1',
                    y='k2',
                    z='k3',
                    color="log_distance",
                    size="log_distance",
                    opacity=1
)
fig.show()
```
