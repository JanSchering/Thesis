import modules

```python
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
import torch as t
import os
from os import path, getcwd
import matplotlib.pyplot as plt
```

load results

```python
df = pd.read_csv(path.join(getcwd(), "data", "results.csv"), usecols=range(1,5))
```

visualize the loss-curves as line graphs

```python
px.defaults.width = 600
fig = px.line()
for DB_val in df["DB"].unique():
    x = df["DA"][df["DB"] == DB_val]
    y = df["log_distance"][df["DB"] == DB_val]
    fig.add_scatter(x=x, y=y, name=f"DB: {DB_val:0.2f}")
    
fig.update_layout(
    title_text='Log-distance for DA values',
    title_x=0.5,
    xaxis_title="DA",
    yaxis_title="log-distance"
)
fig.show()
```

visualize the gradients as line graphs

```python
px.defaults.width = 600
fig = px.line()
for DB_val in df["DB"].unique():
    x = df["DA"][df["DB"] == DB_val]
    y = df["gradient"][df["DB"] == DB_val]
    fig.add_scatter(x=x, y=y, name=f"DB: {DB_val:0.2f}")

fig.add_shape(type='line',
                x0=0,
                y0=0,
                x1=1,
                y1=0,
                line=dict(color='Black', dash="dash"),
                xref='x',
                yref='y'
)
fig.add_shape(type='line',
                x0=0.1,
                y0=-50_000,
                x1=0.1,
                y1=600_000,
                line=dict(color='Gray', dash="dash"),
                xref='x',
                yref='y'
)        
 
fig.update_layout(
    title_text='Gradient measurements',
    title_x=0.5,
    xaxis_title="DA",
    yaxis_title="gradient w.r.t. distance"
)
fig.show()
```

surface plot of the loss

```python
x = df["DA"].unique()
y = df["DB"].unique()
z = df["log_distance"].to_numpy().reshape((len(x), len(y))).T

fig = go.Figure(data=[go.Surface(x=x,y=y,z=z)])

fig.update_layout(scene = dict(
    xaxis_title='DA',
    yaxis_title='DB',
    zaxis_title='log-distance'),
    width=700,
    margin=dict(r=20, b=10, l=10, t=10)
)
fig.show()
```

surface plot of the gradient

```python
x = df["DA"].unique()
y = df["DB"].unique()
z = df["gradient"].to_numpy().reshape((len(x), len(y))).T

fig = go.Figure(data=[go.Surface(x=x,y=y,z=z)])

fig.update_layout(scene = dict(
    xaxis_title='DA',
    yaxis_title='DB',
    zaxis_title='gradient'),
    width=700,
    margin=dict(r=20, b=10, l=10, t=10)
)
fig.show()
```
