```python
def remove_ticklines(ax):
    for tic in ax.xaxis.get_major_ticks():
        tic.tick1line.set_visible(False)
        tic.tick2line.set_visible(False)
        tic.label1.set_visible(False)
        tic.label2.set_visible(False)
    for tic in ax.yaxis.get_major_ticks():
        tic.tick1line.set_visible(False)
        tic.tick2line.set_visible(False)
        tic.label1.set_visible(False)
        tic.label2.set_visible(False)
```

```python
fig = plt.figure(figsize=(10,2), dpi=300)
s_figs = fig.subfigures(1,2)

#frame = s_figs[0].add_axes([0,0,1,1])
def make_number_grid(subfig, position, numbers):
    ax = subfig.add_axes(position)
    remove_ticklines(ax)
    ax.text(-0.2,0.35,numbers[0])
    ax.text(0.8,0.35,numbers[1])
    ax.text(1.8,0.35,numbers[2])
    ax.text(-0.2,1.35,numbers[3])
    ax.text(0.8,1.35,numbers[4])
    ax.text(1.8,1.35,numbers[5])
    ax.text(-0.2,2.35,numbers[6])
    ax.text(0.8,2.35,numbers[7])
    ax.text(1.8,2.35,numbers[8])
    ax.grid(color='black', linewidth=1)
    ax.set_xticks(np.arange(0.5, 3.5, 1))
    ax.set_yticks(np.arange(0.5, 3.5, 1))
    return ax

box_size = [0.3,0.3]
text_pos = [1.,-1.]
init_pos_s = [0.15,0.5]
init_pos_e = [init_pos_s[0],0.1]
offset = 0.2
s_fig = s_figs[1]
    
nums_init_species = ("3","2","2","1","3","2","2","2","4")    
ax_init_species = make_number_grid(
    s_fig,
    [init_pos_s[0],init_pos_s[1],box_size[0],box_size[1]],
    nums_init_species
)
ax_init_species.imshow(np.zeros((3,3)), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)
ax_init_species.text(
    text_pos[0],text_pos[1],
    "Start",
    clip_on=False,
    horizontalalignment="center"
)
ax_init_species.text(
    -2.5,1.25,
    "Species",
    clip_on=False,
    horizontalalignment="center"
)

nums_init_e = ("0","0","0","0","0","0","0","0","0")
ax_init_e = make_number_grid(
    s_fig,
    [init_pos_e[0],init_pos_e[1],box_size[0],box_size[1]],
    nums_init_e
)
ax_init_e.imshow(np.array([int(el) for el in nums_init_e]).reshape(3,3), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)
ax_init_e.text(
    -2.5,1.75,
    "Aux.\nExcitation",
    clip_on=False,
    horizontalalignment="center"
)


nums_mid_species = (
    "2","1","1",
    "1","2","2",
    "1","2","3"
)
ax_mid_species = make_number_grid(
    s_fig,
    [init_pos_s[0]+offset,init_pos_s[1],box_size[0],box_size[1]],
    nums_mid_species
)
ax_mid_species.imshow(np.zeros((3,3)), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)
ax_mid_species.text(
    text_pos[0],text_pos[1],
    "1.) \n Excitation",
    clip_on=False,
    horizontalalignment="center"
)

nums_mid_e = (
"1","1","1",
"0","1","0",
"1","0","1"
)
ax_mid_e = make_number_grid(
    s_fig,
    [init_pos_e[0]+offset,init_pos_e[1],box_size[0],box_size[1]],
    nums_mid_e
)
ax_mid_e.imshow(np.array([int(el) for el in nums_mid_e]).reshape(3,3), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)

nums_trans_species = (
    "2","1","1",
    "1","2","2",
    "1","2","3", 
)
ax_trans_species = make_number_grid(
    s_fig,
    [init_pos_s[0]+offset*2,init_pos_s[1],box_size[0],box_size[1]],
    nums_trans_species
)
ax_trans_species.imshow(np.zeros((3,3)), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)
ax_trans_species.text(
    text_pos[0],text_pos[1],
    "2.) \n Translation",
    clip_on=False,
    horizontalalignment="center"
)

nums_trans_e = (
    "1","0","1",
    "1","1","1",
    "0","1","0"
)
ax_trans_e = make_number_grid(
    s_fig,
    [init_pos_e[0]+offset*2,init_pos_e[1],box_size[0],box_size[1]],
    nums_trans_e
)
ax_trans_e.imshow(np.array([int(el) for el in nums_trans_e]).reshape(3,3), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)

nums_settled_species = (
    "3", "1", "2",
    "2", "3", "3",
    "1", "3", "3"
)
ax_settled_species = make_number_grid(
    s_fig,
    [init_pos_s[0]+offset*3,init_pos_s[1],box_size[0],box_size[1]],
    nums_settled_species
)
ax_settled_species.imshow(np.zeros((3,3)), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)
ax_settled_species.text(
    text_pos[0],text_pos[1],
    "3.) \n Settling",
    clip_on=False,
    horizontalalignment="center"
)

nums_settled_e = ("0", "0", "0", "0", "0", "0", "0", "0", "0")
ax_settled_e = make_number_grid(
    s_fig,
    [init_pos_e[0]+offset*3,init_pos_e[1],box_size[0],box_size[1]],
    nums_settled_e
)
ax_settled_e.imshow(np.zeros((3,3)), cmap="Greys", interpolation="nearest", vmin=0, vmax=1, alpha=0.3)

plt.show()
```
